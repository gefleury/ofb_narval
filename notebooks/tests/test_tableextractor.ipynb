```python
import sys
sys.path.append("../../")    # Add the path to the root directory (where we can find the folder narval/)

%load_ext autoreload
%autoreload 2 

import pandas as pd
import numpy as np
from narval.pdfreader import PDFReader
from narval.tableextractor import TableExtractor
from narval.utils import get_data_dir, FileSystem
```

```python
fs = FileSystem()
data_dir = get_data_dir()
```

```python
file_name = "RPQS_Verneuil-en-Halatte_cp60550_rpqsid_389261_AC_2019.pdf"

pdf_reader = PDFReader(data_dir + "/data/input/pdfs/" + file_name)
```

```python
question_file_path = data_dir + "/data/input/" + "question_keyword_v3.csv"

indicator_list = (fs.read_csv_to_df(question_file_path, usecols=["indic", "competence"])
                .query('competence=="assainissement collectif"')["indic"]
                .unique()
                )
```

```python
page_num = 9

year = 2019

all_tables = pdf_reader.dftables

for df in all_tables[page_num]:
    display(df)

    table = TableExtractor(df, question_file_path, year)
    display(table.df)
    # code_col = table.find_indicator_code_column()
    # val_col = table.find_indicator_value_column()
    # print("Column with indicator codes = ", code_col)
    # print("Column with indicator values = ", val_col)

    if table.is_summary_table():
        sub_df = table.sub_df
        display(sub_df)

    indicator_value_list = []
    for indicator in indicator_list:
        indicator_value_list.append(table.extract_indicator_value(indicator))
    res_dict = {"indicator": indicator_list, "value": indicator_value_list}
    res_df = pd.DataFrame.from_dict(res_dict)
    print("EXTRACTED INDICATORS")
    display(res_df.sort_values(by="indicator"))
```
