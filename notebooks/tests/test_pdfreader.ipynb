## Find the table of content

```python
pdf_list_name = "rpqs_eval_list_1.csv"

# Get the df with pdf names and true toc page numbers
eval_path = "/data/input/" + pdf_list_name
eval_df = fs.read_csv_to_df(data_dir + eval_path, sep=";", usecols=["pdf_name", "toc_page_nb_from_1"])
eval_df["toc_page_nb_from_1"] = eval_df["toc_page_nb_from_1"].apply(ast.literal_eval)

#eval_df = eval_df[eval_df["pdf_name"]=="RPQS_Rully_AC_2021"]


for index, row in eval_df.iterrows():
    pdf_name = row["pdf_name"]
    true_toc = row["toc_page_nb_from_1"]
    pdf_reader = PDFReader(f"{data_dir}/data/input/pdfs/{pdf_name}.pdf")
    toc_pages_numbers = pdf_reader.find_toc_pages()
    print(pdf_name)
    checked = [x-1 for x in true_toc]==toc_pages_numbers
    print(checked)
    if not checked:
        print("true toc_pages_numbers from 0 = ", [x-1 for x in true_toc])
        print("toc_pages_numbers from 0 = ", toc_pages_numbers)
    print("#"*20)
```

## Check detection of RADs

```python
pdf_list_name = "rpqs_eval_list_1+2.csv"

# Get the df with pdf names and true toc page numbers
eval_path = "/data/input/" + pdf_list_name
eval_df = fs.read_csv_to_df(data_dir + eval_path, sep=";", usecols=["pdf_name", "rpqs_rad"])

#eval_df = eval_df[eval_df["pdf_name"]=="RPQS_Rully_AC_2021"]


for index, row in eval_df.iterrows():
    pdf_name = row["pdf_name"]
    rpqs_rad = row["rpqs_rad"]
    pdf_reader = PDFReader(f"{data_dir}/data/input/pdfs/{pdf_name}.pdf")
    checked = pdf_reader.is_rad() == (rpqs_rad=="RAD")
    print(pdf_name)
    print(checked)
```
